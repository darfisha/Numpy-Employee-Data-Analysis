```python
#PART 5 - Flatten, ravel
```

```python
#flatten - converts multi dimensional arrays into one dimension array
b = np.array([[2,3],[1,5],[3,4],[6,8]])
flat = b.flatten()
flat
```

```python
# ravel - returns the flatten array without copy (modyfying it can affect the
# orignal array)
print(b.ravel())
b
```

```python
flat[0] = 99
print("Modified flattened view:", flat)
print("Modified original array:\n", b)
```

```python
flat.base is b
#If it returns True, flat is a view (modifying it will affect b). If False, it's a copy.
```

```python
#PART 8)advance numpy topics

#1) reshape() - change the shape of the array without changing its data
arr = np.arange(6)
arr.reshape(2,3)
```

```python
#2) resize = changes the shape and modifiesb the orignal array if new size > orignal, it fills with garbage values.
arr = np.arange(6)
np.resize(arr, (3,3))

#Garbage values are like random leftover junk in memory that you didn't ask for.
#Random values are like numbers you purposefully ask the computer to generate, and they're following a rule.
```

```python
#3) Transpose or .T = FLIPS ROWS AND COLUMNS
arr = np.array([[0,1,2],[3,4,5]])
arr
print(arr.transpose())
```

```python
a = np.array([1,2,3,4])
print(a.transpose())

#In NumPy, the transpose() method is used to transpose a multi-dimensional array, meaning it switches rows with columns
```

```python
#reshaping
b=a.reshape(-1,1)
print(b.transpose())
```

```python
#PART 9 - BROADCASTING, AXIS BASED OPERATIIONS
a = np.array([[1,2,3],[4,7,8]])
print(np.sum(a,axis=0))
print(np.sum(a,axis=1))
```

```python
#Sorting And Searching
#sort - sort the array
#argsort - return indices of the sorted arrays
#where - return indices where a condition is true
arr = np.array([0,3,2,3,9,5])
np.sort(arr)
np.argsort(arr)
np.where(arr>3)
```

```python
#Set operations
a = np.array([0,1,2,2,3])
b = np.array([3,4,5,6])
np.unique(a)
np.intersect1d(a,b)
np.union1d(a,b)
```

```python
# prompt: #increase the salary of employees in the marketing department by 10% using

import pandas as pd

# Load the dataframe.
df = pd.read_csv("/content/employee.csv")


# Filter for the marketing department and increase their salary by 10%.
df.loc[df['deptname'] == 'Marketing', 'sal'] *= 1.10

# Display the updated dataframe.
df
```

```python
df.sort_values(by = ["sal"] , ascending=False)[["empfirstname", "sal"]].head(3)
```

```python
# Convert 'dateofbirth' to datetime objects for proper comparison
df['dateofbirth'] = pd.to_datetime(df['dateofbirth'])

# Filter employees born before 1990 and update their salary
df.loc[df['dateofbirth'] < pd.Timestamp('1990-01-01'), 'sal'] = 80000

# Display the updated dataframe
df
```
